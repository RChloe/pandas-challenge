## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#unique_username = purchase_data["SN"].unique()
purchase_data_name = purchase_data.drop_duplicates(subset=['SN'])
```

```python
gender_counts = purchase_data_name["Gender"].value_counts().to_frame()
gender_counts.rename(columns={"Gender":"Total Counts"})
percentage_male = str(round(gender_counts.loc["Male"]["Gender"]/player_count*100,2))+"%"
percentage_female = str(round(gender_counts.loc["Female"]["Gender"]/player_count*100,2))+"%"
percentage_other = str(round(gender_counts.loc["Other / Non-Disclosed"]["Gender"]/player_count*100,2))+"%"
percentage = [percentage_male,percentage_female,percentage_other]
gender_counts["Percentage of Players"]=percentage
gender_counts
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
total_purchase_value=pd.DataFrame(purchase_data[["Price","Gender"]].groupby(["Gender"]).sum())
total_purchase_value["Total Purchase Value"]=total_purchase_value["Price"].map('${:,.2f}'.format)
total_purchase_value
```

```python
avg_purchase_price=pd.DataFrame(purchase_data[["Price","Gender"]].groupby(["Gender"]).mean())
avg_purchase_price["Price"]=avg_purchase_price["Price"].map('${:,.2f}'.format)
avg_purchase_price
```

```python
purchase_count=pd.DataFrame(purchase_data[["Price","Gender"]].groupby(["Gender"]).count())
#purchase_count["Price"]=purchase_count["Price"].map('${:,.2f}'.format)
purchase_count
```

```python
male_avg_purch=round(total_purchase_value.loc["Male"]["Price"]/gender_counts.loc["Male"]["Gender"],2)
female_avg_purch=round(total_purchase_value.loc["Female"]["Price"]/gender_counts.loc["Female"]["Gender"],2)
other_avg_purch=round(total_purchase_value.loc["Other / Non-Disclosed"]["Price"]/gender_counts.loc["Other / Non-Disclosed"]["Gender"],2)
avg_purch = [["Male",male_avg_purch],["Female",female_avg_purch],["Other / Non-Disclosed",other_avg_purch]]
df_avg_purch = pd.DataFrame(avg_purch,columns=["Gender","Avg Total Purchase per Person"])
df_avg_purch["Avg Total Purchase per Person"]=df_avg_purch["Avg Total Purchase per Person"].map('${:,.2f}'.format)

df_avg_purch
```

```python
gender_purchases=purchase_count.merge(avg_purchase_price,on="Gender")
gender_purchases=gender_purchases.merge(total_purchase_value,on="Gender")
gender_purchases=gender_purchases.merge(df_avg_purch,on="Gender")
```

```python
gender_purchases=gender_purchases.drop(columns="Price")
gender_purchases.columns=["Gender","Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]
gender_purchases.set_index("Gender")
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9,14,19,24,29,34,39,100]
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
purchase_data["age_group"]=pd.cut(purchase_data["Age"],bins,labels=group_names)
purchase_data_name["age_group"]=pd.cut(purchase_data_name["Age"],bins,labels=group_names)
purchase_data_name.head()
```

```python
age_group = pd.DataFrame(purchase_data_name.groupby("age_group").count())
age_group = pd.DataFrame(age_group["Age"])
age_group
```

```python
percent = []
percent.append(round(age_group["Age"]/player_count*100,2))
percent=pd.DataFrame(percent).T
percent=percent["Age"].map('{0:.2f}%'.format)
```

```python
age_group=age_group.merge(percent,on="age_group")
age_group.columns=["Total Count","Percentage of Players"]
age_group
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
purchase_count2=pd.DataFrame(purchase_data[["Price","age_group"]].groupby(["age_group"]).count())
purchase_count2
```

```python
avg_price2 = pd.DataFrame(purchase_data[["Price","age_group"]].groupby(["age_group"]).mean())
avg_price2["Price"]=avg_price2["Price"].map('${:,.2f}'.format)
avg_price2
```

```python
tot_purch2 = pd.DataFrame(purchase_data[["Price","age_group"]].groupby(["age_group"]).sum())
tot_purch2["Total Purchase Value"]=tot_purch2["Price"].map('${:,.2f}'.format)
tot_purch2
```

```python
avg_tot_purch["Avg Total Purchase per Person"] = pd.DataFrame(tot_purch2["Price"]/age_group["Total Count"])
avg_tot_purch["Avg Total Purchase per Person"]=avg_tot_purch["Avg Total Purchase per Person"].map('${:,.2f}'.format)
avg_tot_purch
```

```python
age_purch=purchase_count2.merge(avg_price2,on="age_group")
age_purch=age_purch.merge(tot_purch2["Total Purchase Value"],on="age_group")
age_purch=age_purch.merge(avg_tot_purch["Avg Total Purchase per Person"],on="age_group")
age_purch.columns=["Purchase Count","Average Purchase Price","Total Purchase Value","Avg Total Purchase per Person"]
age_purch
```
